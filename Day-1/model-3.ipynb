```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('ads-data-3.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Ads Spent (Thousand)')
plt.ylabel('Sales (Millions)')
plt.scatter(df.AdSpend, df.Sales, color='red', marker='.')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['AdSpend']], df.Sales)
```

```python
reg.predict([[3.4]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
0.09075273*3.4+6.001612340669709
```

```python
test=pd.read_csv('test-3.csv')
test
```

```python
p=reg.predict(test)
```

```python
test['predicted-sales']=p
```

```python
test.to_csv('result-3.csv')
```

```python
%matplotlib inline
plt.xlabel('Ads Spent (Thousand)')
plt.ylabel('Sales (Millions)')
plt.scatter(df.AdSpend, df.Sales, color='red', marker='.')
plt.plot(df.AdSpend, reg.predict(df[['AdSpend']]), color='blue')
```
