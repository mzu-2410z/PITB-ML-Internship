```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('ads-data-2.csv')
df
```

```python
%matplotlib inline 
plt.xlabel('Ad Spent (Thousands)')
plt.ylabel('Sales (Millions)')
plt.scatter(df.AdSpend, df.Sales, color='red', marker='.')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['AdSpend']], df.Sales)
```

```python
reg.predict([[3.5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
0.06116442*3.5+6.490752991757189
```

```python
test=pd.read_csv('test-2.csv')
```

```python
test
```

```python
p=reg.predict(test)
```

```python
test['predicted-sales']=p
```

```python
test.to_csv('result-2.csv')
```

```python
%matplotlib inline 
plt.xlabel('Ad Spent (Thousands)')
plt.ylabel('Sales (Millions)')
plt.scatter(df.AdSpend, df.Sales, color='red', marker='.')
plt.plot(df.AdSpend, reg.predict(df[['AdSpend']]), color='blue')
```
