```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('startup-data-3.csv')
df
```

```python
df['R&D Spend']
```

```python
df['Administration']
```

```python
df['Marketing Spend']
```

```python
df['Profit']
```

```python
import math
r=math.floor(df['R&D Spend'].median())
df['R&D Spend']=df['R&D Spend'].fillna(r)
df['R&D Spend']
```

```python
import math
a=math.floor(df['Administration'].median())
df['Administration']=df['Administration'].fillna(a)
df['Administration']
```

```python
import math
m=math.floor(df['Marketing Spend'].median())
df['Marketing Spend']=df['Marketing Spend'].fillna(m)
df['Marketing Spend']
```

```python
import math
a=math.floor(df['Profit'].median())
df['Profit']=df['Profit'].fillna(a)
df['Profit']
```

```python
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['R&D Spend', 'Administration', 'Marketing Spend']], df['Profit'])
```

```python
reg.predict([[35000.60,78000.30,115000.25]])
```

```python
reg.coef_
```
