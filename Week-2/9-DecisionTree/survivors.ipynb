```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
survived = df[df.Survived == 1]
survived.shape
```

```python
dead = df[df.Survived == 0]
dead.shape
```

```python
df.groupby('Survived').mean(numeric_only=True)
```

```python
dummies = pd.get_dummies(df.Sex)
dummies
```

```python
final_dummies = dummies.drop('male', axis = 'columns')
```

```python
final_dummies
```

```python
final_df = df[['Pclass', 'Fare']]
final_df
```

```python
final = pd.concat([final_df, final_dummies], axis = 'columns')
```

```python
final
```

```python
from sklearn.model_selection import train_test_split
```

```python
result = df.Survived
result
```

```python
x_train, x_test, y_train, y_test = train_test_split(final, result, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```
